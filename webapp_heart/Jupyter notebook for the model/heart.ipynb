K Neighbors Classifier

```python
X = df.drop(['output'], axis = 1)
Y = df['output']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=5, p=2, metric="minkowski")
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy: "+ str(accuracy_score(y_test,y_pred)*100) + "%")
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score


from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#Encoding the categorical variable, gende

#Label Encoding.
label = LabelEncoder()
# data['Gender_encoded'] = label.fit_transform(data['sex'])
```

```python
X = df.drop(['output'], axis = 1)
Y = df['output']
```

```python
#Selecting the features to be used to train the model.
train = X.copy()

test = Y.copy()
```

```python
X = np.array(train)

y = np.array(test)
```

```python
def score_function(y_pred, y_test):

    Acc = accuracy_score(y_test, y_pred)
    Pre = precision_score(y_test, y_pred)
    Rec = recall_score(y_test, y_pred)

    data = pd.DataFrame()
    names = ["Accuracy", "Precision", "Recall"]
    values = [Acc, Pre, Rec]
    data["Names"] = names
    data['Scores'] = values

    return data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
#KNearestNeighbors
lr=GaussianNB()
lr.fit(X_train, y_train.ravel())
predictions=lr.predict(X_test)
```

```python
scores = score_function(predictions, y_test)

scores
```

```python
import pickle

with open('heart_disease_naive_bayes.pkl', 'wb') as file:
    pickle.dump(lr, file)
```
